```python
df=pd.read_csv('Cleaned_Viral_Social_Media_Trends.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Post_Date'] = pd.to_datetime(df['Post_Date'])

df['day_of_week'] = df['Post_Date'].dt.day_name()
df['month'] = df['Post_Date'].dt.month
df['is_weekend'] = df['Post_Date'].dt.dayofweek >= 5
```

```python
df = df.drop('Post_Date', axis=1)
```

```python
#Engagement Rate
df['Engagement_rate']=(df['Likes']+df['Shares']+df['Comments'])/df['Views']
df['Engagement_rate']
```

```python
#Hashtag Count
df['Hashtag']=df['Hashtag'].apply(lambda x: len(str (x).split()))
```

```python
df.head()
```

```python
df['Virality_Score']=(df['Likes']*0.4+df['Shares']*0.4+df['Comments']*0.2)/df['Views']
```

```python
df.groupby('Platform')['Engagement_rate'].mean().sort_values(ascending=False)
```

```python
df.groupby('Content_Type')['Engagement_rate'].mean().sort_values(ascending=False)
```

```python
df.groupby('day_of_week')['Engagement_rate'].mean().sort_values(ascending=False)
```

```python
df.head()
```

```python
df.head()
```

```python
df_ = df.drop('Post_ID', axis=1)

```

```python
df = pd.get_dummies(df, columns=[
    'Platform', 'Content_Type', 'Region', 'day_of_week'
], drop_first=True)
```

```python
df = df.drop('Hashtag', axis=1)
```

```python
df = df[df['Likes'] <= df['Views']]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Engagement_Level'] = le.fit_transform(df['Engagement_Level'])
```

```python
df.head()
```

```python
df = df.astype({col: 'int' for col in df.select_dtypes('bool').columns})
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
X = df.drop('Engagement_Level', axis=1)
y = df['Engagement_Level']
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
X = X.astype(int)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
model = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    random_state=42
)
```

```python
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
print("Accuracy:", accuracy_score(ytest, ypred))
print(classification_report(ytest, ypred))
```

```python
import streamlit as st
import pandas as pd
import joblib
```

```python
# Load model + columns
model = joblib.load("viral_model.pkl")
columns = joblib.load("model_columns.pkl")
le = joblib.load("label_encoder.pkl")
```

```python
st.title("📊 AI Viral Content Analyzer")
```

```python
likes = st.number_input("Likes", value=100)
shares = st.number_input("Shares", value=50)
comments = st.number_input("Comments", value=20)
views = st.number_input("Views", value=1000)
```

```python

platform = st.selectbox("Platform", ["Instagram", "YouTube", "Twitter"])
content_type = st.selectbox("Content Type", ["Reel", "Post", "Video"])
region = st.selectbox("Region", ["India", "USA", "UK"])
```

```python
input_dict = {
    'Likes': likes,
    'Shares': shares,
    'Comments': comments,
    'Views': views,
    'Hashtag': 5,
    'Engagement_rate': (likes + shares + comments) / views if views > 0 else 0,
    'Virality_Score': (likes*0.4 + shares*0.4 + comments*0.2) / views if views > 0 else 0
}

input_df = pd.DataFrame([input_dict])
```

```python
# Convert categorical
input_df['Platform'] = platform
input_df['Content_Type'] = content_type
input_df['Region'] = region
```

```python
# One-hot encoding
input_df = pd.get_dummies(input_df)

# Match training columns
input_df = input_df.reindex(columns=columns, fill_value=0)
```

```python
# Prediction
# -------------------------------
if st.button("Predict"):
    pred = model.predict(input_df)
    result = le.inverse_transform(pred)[0]

    if result == "High":
        st.success("🔥 High Engagement (Viral)")
    elif result == "Medium":
        st.warning("⚡ Medium Engagement")
    else:
        st.error("⚠️ Low Engagement")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib

model = joblib.load("viral_model.pkl")
columns = joblib.load("model_columns.pkl")
le = joblib.load("label_encoder.pkl")

st.title("AI Viral Content Analyzer")

likes = st.number_input("Likes", value=100)
shares = st.number_input("Shares", value=50)
comments = st.number_input("Comments", value=20)
views = st.number_input("Views", value=1000)

platform = st.selectbox("Platform", ["Instagram", "YouTube", "Twitter"])
content_type = st.selectbox("Content Type", ["Reel", "Post", "Video"])
region = st.selectbox("Region", ["India", "USA", "UK"])

input_dict = {
    'Likes': likes,
    'Shares': shares,
    'Comments': comments,
    'Views': views,
    'Hashtag': 5,
    'Engagement_rate': (likes + shares + comments) / views if views > 0 else 0,
    'Virality_Score': (likes*0.4 + shares*0.4 + comments*0.2) / views if views > 0 else 0,
    'Platform': platform,
    'Content_Type': content_type,
    'Region': region
}

df = pd.DataFrame([input_dict])
df = pd.get_dummies(df)
df = df.reindex(columns=columns, fill_value=0)

if st.button("Predict"):
    pred = model.predict(df)
    result = le.inverse_transform(pred)[0]
    st.write("Prediction:", result)
```
